```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print("Loaded! Training:", x_train.shape)
plt.imshow(x_train[0], cmap='gray')
plt.title(class_names[y_train[0]])
plt.show()
```

```python
# Normalize and add channels
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Resize using tf.image (safe way)
def resize_and_rgb(x):
    x = tf.expand_dims(x, -1)  # Add channel
    x = tf.image.resize(x, (32, 32))  # Small but enough
    x = tf.repeat(x, 3, axis=-1)  # Fake RGB
    return x

x_train_ready = resize_and_rgb(x_train)
x_test_ready = resize_and_rgb(x_test)

print("Ready shapes:", x_train_ready.shape)  # (60000, 32, 32, 3)
```

```python
from tensorflow.keras.applications import ResNet50, EfficientNetB0, MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential

def make_model(base_class):
    base = base_class(weights='imagenet', include_top=False, input_shape=(32,32,3))
    base.trainable = False
    model = Sequential([
        base,
        GlobalAveragePooling2D(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

epochs = 1
batch_size = 32  # Safe for memory

print("Training ResNet-50...")
resnet = make_model(ResNet50)
resnet.fit(x_train_ready, y_train, epochs=epochs, validation_split=0.2, batch_size=batch_size)
res_acc = resnet.evaluate(x_test_ready, y_test)[1]

print("Training EfficientNetB0...")
effnet = make_model(EfficientNetB0)
effnet.fit(x_train_ready, y_train, epochs=epochs, validation_split=0.2, batch_size=batch_size)
eff_acc = effnet.evaluate(x_test_ready, y_test)[1]

print("Training MobileNetV2...")
mobilenet = make_model(MobileNetV2)
mobilenet.fit(x_train_ready, y_train, epochs=epochs, validation_split=0.2, batch_size=batch_size)
mobile_acc = mobilenet.evaluate(x_test_ready, y_test)[1]

print("\nYOUR FINAL RESULTS:")
print(f"ResNet-50: {res_acc*100:.2f}%")
print(f"EfficientNetB0: {eff_acc*100:.2f}%")
print(f"MobileNetV2: {mobile_acc*100:.2f}%")
```

## Enhanced Version: Using Real Colourful Fashion Dataset

###The original model used Fashion-MNIST (grayscale sketches), which caused misclassifications on real photos (e.g., red gown detected as "Sneaker").

To improve real-world performance, the model was retrained on a colorful product image dataset from Hugging Face ('ashraq/fashion-product-images-small') containing 44,000+ real photos of clothing items.

```python


from datasets import load_dataset
import matplotlib.pyplot as plt

# Load dataset
dataset = load_dataset("ashraq/fashion-product-images-small")

print(f"Loaded! Total images: {len(dataset['train'])}")

# Correct label field is 'articleType' in this dataset
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
for i in range(12):
    row = i // 4
    col = i % 4
    img = dataset['train'][i]['image']
    label = dataset['train'][i]['articleType']  # Yo 'articleType' use gar
    axes[row, col].imshow(img)
    axes[row, col].set_title(label)
    axes[row, col].axis('off')
plt.suptitle("Real Colourful Fashion Products (articleType labels)")
plt.tight_layout()
plt.show()
```

### Prepare Data & Train

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from PIL import Image

# Categories
selected_categories = ['Dresses', 'Tops', 'Skirts', 'Handbags', 'Casual Shoes', 'Formal Shoes']

# Only 3000 images (safe for memory)
filtered = []
for ex in dataset['train']:
    if ex['articleType'] in selected_categories:
        filtered.append(ex)
    if len(filtered) >= 3000:
        break

print(f"Selected {len(filtered)} images")

# Resize to 96x96 (smaller = less memory)
images = []
for ex in filtered:
    img = ex['image'].convert('RGB')
    img = img.resize((96, 96), Image.LANCZOS)
    images.append(np.array(img))

labels = [ex['articleType'] for ex in filtered]

# Encode
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
class_names = encoder.classes_
print("Classes:", class_names)

# Normalize
x = np.array(images) / 255.0
print("Data ready! Shape:", x.shape)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential

# Smaller input
base = ResNet50(weights='imagenet', include_top=False, input_shape=(96,96,3))
base.trainable = False

model = Sequential([
    base,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dense(len(class_names), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train only 8 epochs
model.fit(x, y, epochs=8, validation_split=0.2, batch_size=32)

# Save
model.save('glamify_colorful_final.h5')
print("Model trained and saved!")
```

```python
from PIL import Image


test_img = Image.open('download.jpg')  # If same folder ma chha


test_img = test_img.convert('RGB')
test_img = test_img.resize((96,96))  # Timi ko model size anusar
test_array = np.array(test_img) / 255.0
test_array = np.expand_dims(test_array, axis=0)

pred = model.predict(test_array)
predicted = class_names[np.argmax(pred)]
confidence = np.max(pred) * 100

print(f"Detected: {predicted}")
print(f"Confidence: {confidence:.2f}%")
```
